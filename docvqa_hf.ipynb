```python
from datasets import load_dataset
import PIL.Image
import os
```

```python
ds = load_dataset("lmms-lab/DocVQA", "DocVQA")
```

```python
# Iterate through 100 samples in the dataset and save images locally

images_dir = "images"
os.makedirs(images_dir, exist_ok=True)
for i in range(110):
    sample = ds["validation"][i]
    print(f"Sample {i}:")
    # Save the image to the images subfolder if not already present
    image = sample["image"]
    image_filename = f"sample_{i}.png"
    image_path = os.path.join(images_dir, image_filename)
    if not os.path.exists(image_path):
        image.save(image_path)
    print(f"Saved image to: {image_path}")
```
